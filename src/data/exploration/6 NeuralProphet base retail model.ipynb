```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
sns.plotting_context(context='paper')
sns.set_palette('rainbow')

import pandas as pd
```

```python
transactions = pd.read_csv('../../../data/kaggle/spanish_retail_transactions.csv')
```

```python
transactions['date'] = pd.to_datetime(transactions['date'])
transactions_per_day = transactions.groupby('date')['transactions'].sum().reset_index().rename(columns={'date': 'ds', 'transactions': 'y'})
```

```python
import neuralprophet

m = neuralprophet.NeuralProphet()
m.add_country_holidays(country_name='Spain', lower_window=-4, upper_window=2)

m.fit(transactions_per_day)
```

```python
prediction = pd.DataFrame({'ds': pd.date_range(start='2019-01-01', end='2025-12-31')})
```

```python
prediction['y'] = float('nan')
```

```python
yearly_behaviour_characteristic = m.predict(prediction)
```

```python
yearly_behaviour_characteristic.to_pickle('../../../data/7_NeuralProphet_base_retail_model.pkl')
```
